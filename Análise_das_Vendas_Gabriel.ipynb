```python
# Importar bibliotecas necessárias
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# URL do arquivo CSV
url = "https://github.com/gabrielgpinto/an-lise-de-dados/raw/main/PIPELINE%20DE%20VENDAS%20-%20FEVEREIRO%2073fd3cb875bc4dc9b9522782d4da032a_all.csv"

# Carregar o conjunto de dados
df = pd.read_csv(url)

# Exibir as primeiras linhas do DataFrame
df.head()

# Verificar os tipos de dados das colunas
df.dtypes

# Resumo estatístico do DataFrame
df.describe(include='all')

# Configurações de estilo para os gráficos
sns.set(style="whitegrid")

# Definir o tamanho padrão dos gráficos
plt.figure(figsize=(30, 24))

# Função para adicionar valores totais em cima dos gráficos de barras
def add_total_values(plot, col):
    for p in plot.patches:
        plot.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                      textcoords='offset points')

# Gráfico para variável 'Status'
plt.subplot(2, 3, 1)
status_plot = sns.countplot(x='Status', data=df, hue='Status', palette='viridis', legend=False)
plt.title('Distribuição de Status')
add_total_values(status_plot, 'Status')

# Gráfico para variável 'Adaptado'
plt.subplot(2, 3, 2)
adaptado_plot = sns.countplot(x='Adaptado', data=df, hue='Adaptado', palette='viridis', legend=False)
plt.title('Distribuição de Adaptado')
add_total_values(adaptado_plot, 'Adaptado')

# Gráfico para variável 'Bairro'
plt.subplot(2, 3, 4)
bairro_plot = sns.countplot(x='Bairro', data=df, hue='Bairro', palette='viridis', legend=False)
plt.title('Distribuição de Bairro')
plt.xticks(rotation=45, ha='right')
add_total_values(bairro_plot, 'Bairro')

# Gráfico para variável 'Plano'
plt.subplot(2, 3, 5)
plano_plot = sns.countplot(x='Plano', data=df, hue='Plano', palette='viridis', legend=False)
plt.title('Distribuição de Plano')
plt.xticks(rotation=45, ha='right')
add_total_values(plano_plot, 'Plano')

# Gráfico para variável 'Origem'
plt.subplot(2, 3, 6)
origem_plot = sns.countplot(x='Origem', data=df, hue='Origem', palette='viridis', legend=False)
plt.title('Distribuição de Origem')
plt.xticks(rotation=45, ha='right')
add_total_values(origem_plot, 'Origem')


# Ajustar o layout
plt.tight_layout()
plt.show()

# Gráfico histograma para variável 'Valor'
plt.figure(figsize=(10, 6))
sns.histplot(df['Valor'], kde=True, color='skyblue')
plt.title('Histograma de Valor')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()

# Boxplot para variável 'Valor'
plt.figure(figsize=(8, 5))
sns.boxplot(x='Valor', data=df, color='skyblue')
plt.title('Boxplot de Valor')
plt.show()

# Selecionar apenas as colunas numéricas
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Configurações de estilo para os gráficos
sns.set(style="whitegrid")

```
